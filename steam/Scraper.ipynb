```python
import requests, json, sys, time, re
from bs4 import BeautifulSoup

class steamStructure:
    def __init__(self,name, idNum,numRate, rating, file_size, date_posted, date_updatd, unique, current_subs, current_favs):
        self.name = name # 
        self.idNum = idNum #id number
        self.numRate = numRate
        self.rating = rating 
        self.file_size = file_size
        self.date_posted = date_posted
        self.date_updated = date_updated
        self.unique = unique
        self.current_subs = current_subs
        self.current_favs = current_favs
    def __str__(self):
        return "[\"" + str(self.name) + "\",\""  + str(self.idNum) + "\",\""+ str(self.numRate) + "\",\"" + str(self.rating) + "\",\"" + str(self.file_size) + "\",\"" + str(self.date_posted) + "\",\"" + str(self.date_updated) + "\",\"" + str(self.unique) + "\",\"" + str(self.current_subs) + "\",\"" + str(self.current_favs) +  "\"]"
    def giveJson(self):
        stem =  "\"" +  str(self.idNum) + "\":\"" +"[\"" + str(self.name) + "\",\""  + str(self.idNum) + "\",\""+ str(self.numRate) + "\",\"" + str(self.rating) + "\",\"" + str(self.file_size) + "\",\"" + str(self.date_posted) + "\",\"" + str(self.date_updated) + "\",\"" + str(self.unique) + "\",\"" + str(self.current_subs) + "\",\"" + str(self.current_favs) +  "\"]"
        return stem
    
        
idstring = "?id="
pack = {}
f = open('links.txt','r')
n = open('update.txt','w')

for line in f:
    print(line.rstrip())
    #time.sleep(1)
    #line = "http://steamcommunity.com/sharedfiles/filedetails/?id=214207553&searchtext="
    line =  line.rstrip()
    r = requests.get(line)
    t = r.text
    soup = BeautifulSoup(t, 'html.parser')
    tmp = soup.find_all(class_="workshopItemTitle")
    i=0
    if len(tmp) != 0:
        if(len(tmp[0])!=0):
            if(len(tmp[0].contents[0])!=0):
                name = tmp[0].contents[0]
            else:
                name = -1
        id= line[line.find("?id=")+4: line.find("?id=")+4+line[line.find("?id=")+4:].find("&")]
        table = soup.find_all(class_="detailsStatRight")

    
        size = table[0].contents[0]
        date_posted = table[1].contents[0]
        if len(table) == 3:
            date_updated =  table[2].contents[0]
        else:
            date_updated = -1
        tmp =  soup.find_all(class_="numRatings")
        if len(tmp) != 0:
            tmp = tmp[0].contents[0]
            ratings = tmp[:tmp.find(" ")]
        else:
            ratings = -1
        images = soup.find_all("img")
        for image in images:
            if 'fileRatingDetails'in str(image.parent.attrs):
                stars = image.attrs["src"]
                num = stars.find("-star")
                if num == -1:
                    rating = -1
            #print("rating", stars[num-1:num])
                else:
                    rating = stars[num-1:num]
        table2 = soup.find_all(class_="stats_table")
        split = str(table2)
        tds=[]
        lower = 0
        upper =0
        while lower!= -1:
            if(split[lower:].find("<td>")==-1):
                break
            lower += split[lower:].find("<td>")
            upper = lower +  split[lower:].find("</td>")
            tds.append(split[lower+4:upper])
            if lower != -1:
                lower +=1

        unique = tds[0]
        current_subs = tds[2]
        current_favs = tds[4]
        pack[id] = steamStructure(name, id, ratings, rating, size, date_posted, date_updated, unique, current_subs, current_favs)
        print(pack[id])
        n.write(str(i) + "\n")


output = open("output.txt",'w')
output.write("{")
for mod in pack.keys():
    output.write(pack[mod].giveJson())
output.write("}\n")
output.close()
print("done")
```
