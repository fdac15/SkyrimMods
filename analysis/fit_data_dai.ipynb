```r
data = read.csv("fit_data_dai.csv")
```

```r
data = data[,colSums(data) >0 ]
dim(data)
```

```r
sum(is.na(data))
```

```r
summary(data)
```

```r
# Define a function
hiCor <- function(x, level){
  res <- cor(x,method="spearman");
  res1 <- res; res1[res<0] <- -res[res < 0];
  for (i in 1:dim(x)[2]){
    res1[i,i] <- 0;
  }
  sel <- apply(res1,1,max) > level;
  res[sel,sel];
}
hiCor(data,.7)
```
