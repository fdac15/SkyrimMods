#Analysis on Skyrim Data

```r
f = read.csv("skyrim.csv");

vars = cbind( f$likes, f$downloads, f$time, f$unique_dl, f$views);

dimnames(vars)[[2]] <- c("likes", "downloads", "time", "unique_dl", "views");

data <- data.frame(vars);
```

```r
summary(data);
```

```r
cor(vars, use="pairwise.complete.obs", method="spearman"); 
```

```r
res<-prcomp(vars, retx=F,scale=T, na.action = na.omit)$rotation[,1:3];
resAbs <- res;
resAbs[res<0] <- -res[res<0];
for (i in 1:3)
  print(t(res[resAbs[,i]>.3,i,drop=FALSE]));
```

```r
plot(1:5,cumsum(prcomp(vars, retx=F,scale=T)$sdev^2)/sum(prcomp(vars, retx=F,scale=T)$sdev^2),ylim=c(0,1),xlab="Number of coponents",ylab="Fraction of variance");
```
